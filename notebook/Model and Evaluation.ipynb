```python
import pandas as pd
from sklearn.model_selection import train_test_split,RandomizedSearchCV,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,make_scorer
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix,classification_report
```

```python
data = pd.read_csv("Data/Transformation.csv")
# data.drop(columns=["loan_status.1"],inplace=True)
```

```python
x = data.drop(columns=["loan_status"])
Y = data.iloc[:,27]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,Y,test_size=0.2,random_state=23)
```

```python
type(y_test)
```

<h4>Logistic Regression</h4>

```python
LR_model = LogisticRegression()
LR = LR_model.fit(x_train,y_train)
LR
```

```python
Test = LR.predict(x_test)
Test
```

```python
data.columns
```

```python
accuracy = accuracy_score(y_test,Test)
precision_ = precision_score(y_test,Test)
print(accuracy)
print(precision_)
```

```python
#TUNING 

params = {
    'C': [0.01, 0.1, 1, 10],
    "penalty":["l2","l1"],
    "class_weight":[None,"balanced"],
    "solver":["liblinear"],
    "max_iter":[100,200,300]
}
RSV = RandomizedSearchCV(LR_model,params,cv=5,scoring='accuracy')
RSV = RSV.fit(x_train,y_train)
RSV
```

```python
RSV.best_params_
```

```python
LR_tune = LogisticRegression(solver='liblinear',penalty='l1',max_iter=100,class_weight=None,C= 1)
LR_tune = LR_tune.fit(x_train,y_train)
LR_tune
```

```python
Test_tune = LR_tune.predict(x_test)
Test_tune
```

```python
accuracy = accuracy_score(y_test,Test_tune)
precision_ = precision_score(y_test,Test_tune)
f1 = f1_score(y_test,Test_tune)
print(f1)
print(accuracy)
print(precision_)
```

<h4>SVC</h4>

```python
SVC_model = SVC(kernel="poly")
svc = SVC_model.fit(x_train,y_train)
svc
```

```python
predict = svc.predict(x_test)
predict
```

```python
print(accuracy_score(y_test,predict))
print(precision_score(y_test,predict))
print(f1_score(y_test,predict))
```

```python
f1_ = make_scorer(f1_score)
params = {
    'C': np.logspace(-2, 2, 10),
    "kernel":["rbf","poly","linear"],
    "gamma":[0.1,1,10,"scale","auto"],
    "degree":[2,3,5]
}
svc_tune = RandomizedSearchCV(SVC_model,params,cv=5,n_iter=10,scoring=f1_)
svc_tune.fit(x_train,y_train)
svc_tune
```

```python
best_model = svc_tune.best_estimator_
pred = best_model.predict(x_test)
print(svc_tune.best_score_)
print("f1 score:",f1_score(y_test,pred,pos_label="Y"))
print("accuracy:",accuracy_score(y_test,pred))
print("precision:",precision_score(y_test,pred,pos_label="Y"))
```

<H4>KNN</H4>

```python
KNN = KNeighborsClassifier(n_neighbors=10)
KNN = KNN.fit(x_train,y_train)
KNN
```

```python
predict = KNN.predict(x_test)
predict
```

```python
print(f1_score(y_test,predict))
print(accuracy_score(y_test,predict))
print(precision_score(y_test,predict))
```

```python
# Tune
params = {
    "n_neighbors":[5,10,12,15],
    "weights":["uniform","distance"],
    "algorithm":["kd_tree","ball_tree","brute","auto"],
    "leaf_size":[10,15,30,50,100]
}
f1_ = make_scorer(f1_score,pos_label="Y")
knn_tune = RandomizedSearchCV(KNN,param_distributions=params,scoring=f1_,n_iter=10,cv=5)
knn_tune.fit(x_train,y_train)
knn_tune
```

```python
best_model = knn_tune.best_estimator_
predict = best_model.predict(x_test)
print(knn_tune.best_score_)
print("f1_score:",f1_score(y_test,predict,pos_label="Y"))
print(accuracy_score(y_test,predict))
print(precision_score(y_test,predict,pos_label="Y"))
```

## Decision Tree

```python
New_data = pd.read_csv("Data/transformation_org.csv")
New_data.head()
```

```python
X = New_data.drop(columns=["loan_status"])
y = New_data.iloc[:,27]
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.20,random_state=33)
```

```python
dtc = DecisionTreeClassifier(criterion='entropy',min_samples_split=2,min_samples_leaf=8,max_depth=5)
dtc.fit(X_train,Y_train)
dtc
```

```python
predict = dtc.predict(X_test)
print("f1_score:",f1_score(Y_test,predict))
print(accuracy_score(Y_test,predict))
print(precision_score(Y_test,predict))
```

```python
#tuning
params = {
    "criterion":["gini","entropy","log_loss"],
    "splitter":["best","random"],
    "min_samples_split":[2,4,8,10,15],
    "min_samples_leaf":[2,5,8,10],
    "max_depth":[2,5,8,10]
}
f1_ = make_scorer(f1_score)
dtc_tune = RandomizedSearchCV(dtc,param_distributions=params,scoring=f1_,cv=5,n_iter=10)
dtc_tune.fit(X_train,Y_train)
dtc_tune
```

```python
best_model = dtc_tune.best_estimator_
predict = best_model.predict(X_test)
print("f1_score:",f1_score(Y_test,predict))
print(accuracy_score(Y_test,predict))
print(precision_score(Y_test,predict))
```
